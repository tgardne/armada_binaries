```python
## Query Kervella -- Table 2
print('HD HIP Name Plx e_Plx')
print('--'*10)

for target in target_list:
    
    kervella_query_2 = Vizier.query_object("HD %s"%target,catalog='J/A+A/657/A7/tablea2')

    hip_numbers = np.asarray(kervella_query_2[0]['HIP'])
    names = np.asarray(kervella_query_2[0]['Name'])
    plx = np.asarray(kervella_query_2[0]['Plx'])
    e_plx = np.asarray(kervella_query_2[0]['e_Plx'])
    
    for hh,nn,pp,ee in zip(hip_numbers,names,plx,e_plx):
        print(target,hh,nn,pp,ee)
    
    print('--'*10)
```
