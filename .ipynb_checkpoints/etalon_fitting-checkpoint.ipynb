```python
## the etalon model
def mas2rad(mas):
    rad=mas/1000.
    rad=rad/3600.
    rad=rad*np.pi/180.
    return(rad)

def etalon(u,v,ratio,ud,delta_s,lamb,bw):
    lamb2 = lamb*1e6
    n = np.sqrt(0.6961663*lamb2**2/(lamb2**2-0.0684043**2)+0.4079426*lamb2**2/(lamb2**2-0.1162414**2)+0.8974794*lamb2**2/(lamb2**2-9.896161**2)+1)

    #alpha = ratio + slope2*(lamb-1.4e-6)

    secondary_flux = 1/(1+ratio+1/ratio)
    primary_flux = 1/(1+1/ratio+1/(ratio*ratio))
    tertiary_flux = 1 - primary_flux - secondary_flux

    diameter = mas2rad(ud)
    x = np.pi*diameter*np.sqrt(u**2+v**2)
    
    ## bessel 1st order
    f1 = 2*special.jn(1,x)/x
    ## bandwidth smearing
    vis_bw = np.sinc(np.pi*bw*(n*delta_s/lamb))
        
    complex_vis = primary_flux*f1 + vis_bw*secondary_flux*f1*np.exp(-2*np.pi*1j*n*delta_s/lamb) + vis_bw*tertiary_flux*f1*np.exp(-4*np.pi*1j*n*delta_s/lamb)

    return complex_vis
```

```python
## function which returns complex vis given sep, pa, flux ratio, HA, dec, UD1, UD2, wavelength
def cvis_model(params, delta_s, u, v, wl):
    ratio = params['ratio']
    ud = params['ud']
    bw = params['bw']
    ul=u/wl
    vl=v/wl  
    vis=etalon(ul,vl,ratio,ud,delta_s,wl,bw)
    return vis

## for closure phase
def cvis_model_cp(params, delta_s1, delta_s2, delta_s3, u, v, wl):  
    ratio = params['ratio']
    ud = params['ud']
    bw = params['bw']
    ul=u/wl
    vl=v/wl
    delta_s=np.array([delta_s1,delta_s2,delta_s3])
    vis=etalon(ul,vl,ratio,ud,delta_s,wl,bw)
    return vis
```

## Closure Phase

```python
## the etalon model
def mas2rad(mas):
    rad=mas/1000.
    rad=rad/3600.
    rad=rad*np.pi/180.
    return(rad)

def etalon(u,v,ratio,ud,delta_s,lamb,s1,s2):
    
    ## first need n at lambda - using a slope
    n = 1.45090 + s1*(lamb-1e-6)
    
    ## now need spectral slope in flux ratio
    alpha = ratio + s2*(lamb-1.4e-6)
    
    secondary_flux = 1/(1+alpha)
    primary_flux = 1-secondary_flux
    
    diameter = mas2rad(ud)
    x = np.pi*diameter*np.sqrt(u**2+v**2)
    
    ## bessel 1st order
    f1 = 2*special.jn(1,x)/x
        
    complex_vis = primary_flux*f1 + secondary_flux*f1*np.exp(-2*np.pi*1j*n*delta_s/lamb)
    #complex_vis = f1 + ratio*np.exp(-2*np.pi*1j*1.5*delta_s/lamb)

    return complex_vis
```

```python
## function which returns complex vis given sep, pa, flux ratio, HA, dec, UD1, UD2, wavelength
def cvis_model(params, delta_s1, delta_s2, delta_s3, u, v, wl):
    
    ratio = params['ratio']
    ud = params['ud']
    slope1 = params['slope1']
    slope2 = params['slope2']

    ul=u/wl
    vl=v/wl

    delta_s=np.array([delta_s1,delta_s2,delta_s3])
    
    vis=etalon(ul,vl,ratio,ud,delta_s,wl,slope1,slope2)
    return vis

## function which returns residual of model and data to be minimized
def cp_minimizer(params,cp,cp_err,u_coords,v_coords,wl,tel):
    
    delta_s1 = params['delta_s1']
    delta_s2 = params['delta_s2']
    delta_s3 = params['delta_s3']
    
    model = []
    for item1,item2,item3 in zip(u_coords,v_coords,wl):
        phase_arr=[]
        for w in item3:
            complex_vis = cvis_model(params,delta_s1,delta_s2,delta_s3,item1,item2,w)
            phase = (np.angle(complex_vis[0])+np.angle(complex_vis[1])+np.angle(complex_vis[2]))
            phase_arr.append(phase)
        model.append(phase_arr)
    model=np.array(model)
    
    ## need to do an "angle difference"
    data = cp*np.pi/180
    err = cp_err*np.pi/180

    diff = np.arctan2(np.sin(data-model),np.cos(data-model))
    return diff/err
```

```python
## Ask the user which file contains the closure phases
dir='/Volumes/tg_backup/MIRCX_data/reduced/2018Sep20/oifits/etalon/calibrated/'
target_id='HD_5448'

t3phi = []
t3phierr = []
u_coords = []
v_coords = []
eff_wave = []
time_obs=[]
tels = []

for file in os.listdir(dir):
    if file.endswith(".fits"):
        
        filename = os.path.join(dir, file)
        oifile = oifits.open(filename,quiet=True)
        oi_t3 = oifile.t3
        
        for i in eachindex(oi_t3):
            if str(oi_t3[i].target.target)==target_id:

                t3phi.append(oi_t3[i].t3phi.data)
                t3phierr.append(oi_t3[i].t3phierr)
                eff_wave.append(oi_t3[i].wavelength.eff_wave)
                time_obs.append(oi_t3[i].timeobs)
                tels.append([a.sta_name for a in oi_t3[i].station])
                
                u1coord = oi_t3[i].u1coord
                v1coord = oi_t3[i].v1coord
                u2coord = oi_t3[i].u2coord
                v2coord = oi_t3[i].v2coord
                u3coord = -u1coord - u2coord
                v3coord = -v1coord - v2coord
                
                u_coords.append([u1coord,u2coord,u3coord])
                v_coords.append([v1coord,v2coord,v3coord])

t3phi = np.array(t3phi)
t3phierr = np.array(t3phierr)
u_coords = np.array(u_coords)
v_coords = np.array(v_coords)
eff_wave = np.array(eff_wave)
time_obs = np.array(time_obs)
tels = np.array(tels)
```

```python
## pick a triangle to fit
triangle = tels[5]

t3phi_new = []
t3phierr_new = []
ucoord_new = []
vcoord_new = []

for t3,t3err,uc,vc,bl in zip(t3phi,t3phierr,u_coords,v_coords,tels):
    if str(triangle)==str(bl):
        t3phi_new.append(t3)
        t3phierr_new.append(t3err)
        ucoord_new.append(uc)
        vcoord_new.append(vc)
        
t3phi_new = np.array(t3phi_new)
t3phierr_new = np.array(t3phierr_new)
ucoord_new = np.array(ucoord_new)
vcoord_new = np.array(vcoord_new)
```

```python
###########
## Now do least-squares minimization -- using lmfit python package
###########

a1 = 4.0
a2 = 0.67

slope1_guess = -0.01227e-6
slope2_guess = 0

#create a set of Parameters, choose starting value and range for search
params = Parameters()

## star params
params.add('ratio', value= a1)#min=1.0, max=100.0)
params.add('ud',   value= a2, vary=False)#min=0.1,max=5.0)
params.add('slope1', value=slope1_guess)
params.add('slope2', value=slope2_guess,vary=False)

## etalon params
params.add('delta_s1',value=12e-6)
params.add('delta_s2',value=36e-6)
params.add('delta_s3',value=48e-6)
```

```python
#do fit, minimizer uses LM for least square fitting of model to data
minner = Minimizer(cp_minimizer, params, fcn_args=(t3phi_new,t3phierr_new,ucoord_new,vcoord_new,eff_wave,tels),nan_policy='omit')
result = minner.minimize()
report_fit(result)
```

```python
chi_sq = result.redchi
ratio_result = result.params['ratio'].value
ud_result = result.params['ud'].value
slope1_result = result.params['slope1'].value
slope2_result = result.params['slope2'].value

delta_s1_result = result.params['delta_s1'].value
delta_s2_result = result.params['delta_s2'].value
delta_s3_result = result.params['delta_s3'].value

## plot data with best fit model
best_params = Parameters()
best_params.add('ratio',value=ratio_result)
best_params.add('ud',value=ud_result)
best_params.add('slope1', value=slope1_result)
best_params.add('slope2', value=slope2_result)

best_params.add('delta_s1',value=delta_s1_result)
best_params.add('delta_s2',value=delta_s2_result)
best_params.add('delta_s3',value=delta_s3_result)
```

```python
model = []
for item1,item2,item3,item4 in zip(u_coords,v_coords,eff_wave,tels):
    phase_arr=[]
    for w in item3:
        complex_vis = cvis_model(best_params,delta_s1_result,delta_s2_result,delta_s3_result,item1, item2, w)
        phase = (np.angle(complex_vis[0])+np.angle(complex_vis[1])+np.angle(complex_vis[2]))*180/np.pi
        phase_arr.append(phase)
    model.append(phase_arr)
model=np.array(model)
```

```python
## the etalon model
def mas2rad(mas):
    rad=mas/1000.
    rad=rad/3600.
    rad=rad*np.pi/180.
    return(rad)

def etalon(u,v,ratio,ud,delta_s,lamb,s1,s2):
    
    ## first need n at lambda - using a slope
    n = 1.45090 + s1*(lamb-1e-6)
    
    ## now need spectral slope in flux ratio
    alpha = ratio + s2*(lamb-1.4e-6)
    
    secondary_flux = 1/(1+alpha)
    primary_flux = 1-secondary_flux
    
    diameter = mas2rad(ud)
    x = np.pi*diameter*np.sqrt(u**2+v**2)
    
    ## bessel 1st order
    f1 = 2*special.jn(1,x)/x
        
    complex_vis = primary_flux*f1 + secondary_flux*f1*np.exp(-2*np.pi*1j*n*delta_s/lamb)
    #complex_vis = f1 + ratio*np.exp(-2*np.pi*1j*1.5*delta_s/lamb)

    return complex_vis
```

```python
## function which returns complex vis given sep, pa, flux ratio, HA, dec, UD1, UD2, wavelength
def cvis_model(params, u, v, wl):
    
    delta_s = params['delta_s']
    ratio = params['ratio']
    ud = params['ud']
    s1 = params['s1']
    s2 = params['s2']

    ul=u/wl
    vl=v/wl
    
    vis=etalon(ul,vl,ratio,ud,delta_s,wl,s1,s2)
    return vis

## function which returns residual of model and data to be minimized
def visphi_minimizer(params,vis,viserr,u_coords,v_coords,wl):
    
    model = []
    for item1,item2,item3 in zip(u_coords,v_coords,wl):
        vis_arr=[]
        for w in item3:
            complex_vis = cvis_model(params,item1, item2, w)
            visibility = np.angle(complex_vis)
            vis_arr.append(visibility)
        model.append(vis_arr)
    model=np.array(model)

    ## need to do an "angle difference"
    data = vis*np.pi/180
    err = viserr*np.pi/180

    diff = np.arctan2(np.sin(data-model),np.cos(data-model))
    return diff/err
```

```python
## Ask the user which file contains the closure phases
dir='/Volumes/tg_backup/MIRCX_data/reduced/2018Oct22/oifits_ncoh5_nbs1_ncs1/etalon/calibrated/'
target_id='HD_886'

visphi=[]
visphierr=[]
vistels=[]
ucoord=[]
vcoord=[]
eff_wave=[]
for file in os.listdir(dir):
    if file.endswith(".fits"):
        
        filename = os.path.join(dir, file)
        oifile = oifits.open(filename,quiet=True)
        oi_vis = oifile.vis
        
        for i in eachindex(oi_vis):
            if str(oi_vis[i].target.target)==target_id:
                visphi.append(oi_vis[i].visphi.data)
                visphierr.append(oi_vis[i].visphierr.data)
                eff_wave.append(oi_vis[i].wavelength.eff_wave)
                vistels.append([a.sta_name for a in oi_vis[i].station])
                ucoord.append(oi_vis[i].ucoord)
                vcoord.append(oi_vis[i].vcoord)
visphi = np.array(visphi)
visphierr = np.array(visphierr)
ucoord = np.array(ucoord)
vcoord = np.array(vcoord)
eff_wave = np.array(eff_wave)
vistels=np.array(vistels)
```

```python
## pick a baseline to fit
baseline = ['E1','W2']

visphi_new = []
visphierr_new = []
ucoord_new = []
vcoord_new = []

for vphi,vphierr,uc,vc,bl in zip(visphi,visphierr,ucoord,vcoord,vistels):
    if str(baseline[0])==str(bl[0]) and str(baseline[1])==str(bl[1]):
        visphi_new.append(np.gradient(vphi))
        visphierr_new.append(vphierr)
        ucoord_new.append(uc)
        vcoord_new.append(vc)
        
visphi_new = np.array(visphi_new)
visphierr_new = np.array(visphierr_new)
ucoord_new = np.array(ucoord_new)
vcoord_new = np.array(vcoord_new)
```

```python
visphi_new.shape
```

```python
###########
## Now do least-squares minimization -- using lmfit python package
###########

a1 = 4.0
a2 = 0.4
offset_guess = 1
delta_s_guess = 48e-6
s1_guess = -0.01227e-6
s2_guess = 0

#create a set of Parameters, choose starting value and range for search
params = Parameters()
## star params
params.add('ratio', value= a1)
params.add('ud',   value= a2, vary=False)
params.add('offset', value=offset_guess)
params.add('s1', value=s1_guess)
params.add('s2', value=s2_guess)
## etalon params
params.add('delta_s',value=delta_s_guess)#,min=delta_s_guess-5,max=delta_s_guess+5)

minner = Minimizer(visphi_minimizer, params, fcn_args=(visphi_new,visphierr_new,ucoord_new,vcoord_new,eff_wave),nan_policy='omit')
result = minner.minimize()
report_fit(result)

chi_sq = result.redchi
ratio_result = result.params['ratio'].value
ud_result = result.params['ud'].value
delta_s_result = result.params['delta_s'].value
offset_result = result.params['offset'].value
s1_result = result.params['s1'].value
s2_result = result.params['s2'].value
```

```python
## plot data with best fit model
best_params = Parameters()
best_params.add('ratio',value=ratio_result)
best_params.add('ud',value=ud_result)
best_params.add('delta_s',value=delta_s_result)
best_params.add('offset',value=offset_result)
best_params.add('s1',value=s1_result)
best_params.add('s2',value=s2_result)

model = []
for item1,item2,item3 in zip(ucoord_new,vcoord_new,eff_wave):         
    vis_arr=[]
    for w in item3:
        complex_vis = cvis_model(best_params,item1, item2, w)
        visibility = np.angle(complex_vis)*180/np.pi
        vis_arr.append(visibility)
    model.append(vis_arr)
model=np.array(model)

## plot vis2 data
for v,verr,uc,vc,m in zip(visphi_new,visphierr_new,ucoord_new,vcoord_new,model):
    
    offset = best_params['offset']
    
    x=eff_wave[0]
    #v=v[::-1]
    
    plt.errorbar(x,v,yerr=verr,fmt='o-',zorder=1)
    plt.plot(x,m/offset,'+--',color='r',zorder=2)

plt.title(baseline)
plt.xlabel('Wavelength (m)')
plt.ylabel('VisPhi')
plt.show()
```
